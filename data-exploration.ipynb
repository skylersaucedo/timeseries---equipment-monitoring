```python
"""
use this to explore data and arundo API's various detectors
"""
import os
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
import numpy as np
import pandas as pd
import adtk

# use Arundo's Detector library to validate data. Look for nulls, repeat time stamps.

from adtk.data import validate_series

df = pd.read_csv('sensor_data.csv', index_col="datetime", parse_dates=True)
df = validate_series(df)
df
```

```python
# look at each col
axes = df[df.columns].plot(linewidth=2, figsize=(25, 45), subplots=True, fontsize=23)
```

```python
# run Autoreggresion AD on power and flow to see if we can detect anom early on...

from adtk.detector import AutoregressionAD
from adtk.visualization import plot

cyclical_data = df[['system power', 'pump flow']]

# these are good, predicts anom early on
#n_steps = 90
#step_size = 90

# seems to center around event of breakdown...
#n_steps = 120
#step_size = 120

n_steps = 75 # include half cycle of pump
step_size = 158 # approximate cycle of pump


autoregression_ad = AutoregressionAD(n_steps=n_steps, step_size=step_size, c=3.0)
anomalies = autoregression_ad.fit_detect(cyclical_data)
plot(cyclical_data, anomaly=anomalies, ts_markersize=1, anomaly_color='red', anomaly_tag="marker", anomaly_markersize=3);

# everything before green line is nominal, failure occurs at black
plt.axvline(pd.to_datetime('2019-11-06 13:00:00'), color='g', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2019-11-10 16:00:00'), color='black', linestyle='--', lw=2)
plt.show()
```

```python
pressure_data = df[['lube tank pressure', 'lube gearbox pressure']]

#n_steps = 90
#step_size = 90

n_steps = 75 # include half cycle of pump
step_size = 158 # approximate cycle of pump


autoregression_ad = AutoregressionAD(n_steps=n_steps, step_size=step_size, c=3.0)
anomalies = autoregression_ad.fit_detect(pressure_data)
plot(pressure_data, anomaly=anomalies, ts_markersize=1, anomaly_color='red', anomaly_tag="marker", anomaly_markersize=3);

# everything before green line is nominal, failure occurs at black
plt.axvline(pd.to_datetime('2019-11-06 13:00:00'), color='g', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2019-11-10 16:00:00'), color='black', linestyle='--', lw=2)
plt.show()

```

```python
temp_data = df[['lube gearbox temperature']]

#n_steps = 90
#step_size = 90

n_steps = 75 # include half cycle of pump
step_size = 158 # approximate cycle of pump


autoregression_ad = AutoregressionAD(n_steps=n_steps, step_size=step_size, c=3.0)
anomalies = autoregression_ad.fit_detect(temp_data)
plot(temp_data, anomaly=anomalies, ts_markersize=1, anomaly_color='red', anomaly_tag="marker", anomaly_markersize=3);

# everything before green line is nominal, failure occurs at black
plt.axvline(pd.to_datetime('2019-11-06 13:00:00'), color='g', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2019-11-10 16:00:00'), color='black', linestyle='--', lw=2)
plt.show()
```

```python
# min cluster detector: pressure_data

n_clusters = 5 # 5 is good

from adtk.detector import MinClusterDetector
from sklearn.cluster import KMeans
min_cluster_detector = MinClusterDetector(KMeans(n_clusters=n_clusters))
anomalies = min_cluster_detector.fit_detect(pressure_data)
plot(pressure_data, anomaly=anomalies, ts_linewidth=1, ts_markersize=3, anomaly_color='red', anomaly_alpha=0.3, curve_group='all')

# everything before green line is nominal, failure occurs at black
plt.axvline(pd.to_datetime('2019-11-06 13:00:00'), color='g', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2019-11-10 16:00:00'), color='black', linestyle='--', lw=2)
plt.show()
```

```python
# min cluster detector cyclical_data

n_clusters = 30 # 5 is good

from adtk.detector import MinClusterDetector
from sklearn.cluster import KMeans
min_cluster_detector = MinClusterDetector(KMeans(n_clusters=n_clusters))
anomalies = min_cluster_detector.fit_detect(cyclical_data)
plot(cyclical_data, anomaly=anomalies, ts_linewidth=1, ts_markersize=3, anomaly_color='red', anomaly_alpha=0.3, curve_group='all')

# everything before green line is nominal, failure occurs at black
plt.axvline(pd.to_datetime('2019-11-06 13:00:00'), color='g', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2019-11-10 16:00:00'), color='black', linestyle='--', lw=2)
plt.show()
```

```python
temp_data = df[['lube gearbox temperature']]


from adtk.detector import QuantileAD
quantile_ad = QuantileAD(high=0.981, low=0.01) # slightly lowered sensitivity
anomalies = quantile_ad.fit_detect(temp_data)

plot(temp_data, anomaly=anomalies, ts_linewidth=1, ts_markersize=3, anomaly_markersize=5, anomaly_color='red', anomaly_tag="marker");


# everything before green line is nominal, failure occurs at black
plt.axvline(pd.to_datetime('2019-11-06 13:00:00'), color='g', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2019-11-10 16:00:00'), color='black', linestyle='--', lw=2)
plt.show()
```

```python
# temp change in variation 

temp_data = df[['lube gearbox temperature']]

# use quartile detector 

from adtk.detector import InterQuartileRangeAD
iqr_ad = InterQuartileRangeAD(c=10) # c = 4 works
anomalies = iqr_ad.fit_detect(temp_data)

plot(temp_data, anomaly=anomalies, ts_linewidth=1, ts_markersize=3, anomaly_markersize=5, anomaly_color='red', anomaly_tag="marker");

# everything before green line is nominal, failure occurs at black
plt.axvline(pd.to_datetime('2019-11-06 13:00:00'), color='g', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2019-11-10 16:00:00'), color='black', linestyle='--', lw=2)
plt.show()
```

```python
# level shift AD with temp data

from adtk.detector import LevelShiftAD
level_shift_ad = LevelShiftAD(c=10.0, side='both', window=5)
anomalies = level_shift_ad.fit_detect(temp_data)
plot(temp_data, anomaly=anomalies, anomaly_color='red')

# everything before green line is nominal, failure occurs at black
plt.axvline(pd.to_datetime('2019-11-06 13:00:00'), color='g', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2019-11-10 16:00:00'), color='black', linestyle='--', lw=2)
plt.show()
```

```python
# see if volatility shift on temp data

from adtk.detector import VolatilityShiftAD

n_steps = 150
volatility_shift_ad = VolatilityShiftAD(c=10.0, side='positive', window=n_steps, agg='std')
anomalies = volatility_shift_ad.fit_detect(temp_data)
plot(temp_data, anomaly=anomalies, anomaly_color='red')

# everything before green line is nominal, failure occurs at black
plt.axvline(pd.to_datetime('2019-11-06 13:00:00'), color='g', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2019-11-10 16:00:00'), color='black', linestyle='--', lw=2)
plt.show()
```

```python
# see if volatility shift can be used on pump data


from adtk.detector import VolatilityShiftAD

lube_press = df[['lube gearbox pressure']]


n_steps = 75
volatility_shift_ad = VolatilityShiftAD(c=3.0, side='positive', window=n_steps, agg='std')
anomalies = volatility_shift_ad.fit_detect(lube_press)
plot(lube_press, anomaly=anomalies, anomaly_color='red');

# everything before green line is nominal, failure occurs at black
plt.axvline(pd.to_datetime('2019-11-06 13:00:00'), color='g', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2019-11-10 16:00:00'), color='black', linestyle='--', lw=2)
plt.show()
```

```python
from adtk.detector import OutlierDetector
from sklearn.neighbors import LocalOutlierFactor
outlier_detector = OutlierDetector(LocalOutlierFactor(contamination=0.01))
anomalies = outlier_detector.fit_detect(pressure_data)
plot(pressure_data, anomaly=anomalies, ts_linewidth=1, ts_markersize=3, anomaly_color='red', anomaly_alpha=0.3, curve_group='all');
```

```python

# try PCA on pressure data

from adtk.detector import PcaAD
pca_ad = PcaAD(k=2)
anomalies = pca_ad.fit_detect(pressure_data)
plot(pressure_data, anomaly=anomalies, ts_linewidth=1, ts_markersize=3, anomaly_color='red', anomaly_alpha=0.3, curve_group='all')

# everything before green line is nominal, failure occurs at black
plt.axvline(pd.to_datetime('2019-11-06 13:00:00'), color='g', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2019-11-10 16:00:00'), color='black', linestyle='--', lw=2)
plt.show()
```
